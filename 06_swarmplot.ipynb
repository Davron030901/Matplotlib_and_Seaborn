# Seaborn: swarmplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

## Intro Visuals

```python
sns.set_style('white')
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
```

```python
blue, orange, green, red = sns.color_palette()[:4]
```

```python
pts = np.array([1, 1, 1, 1, 1])  # , 2, 2, 3, 4, 4, 4, 4, 5, 5, 7]

# Plot
plt.figure(figsize=(4, 4))
sns.swarmplot(x=pts, color='green', size=7, linewidth=1, edgecolor='black')  # Reduced size
sns.despine(left=True)
plt.xlim(0, 9)  # Adjusted x-axis range if needed
plt.tight_layout()
```

## Basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.swarmplot(cars.horsepower)
```

```python
sns.swarmplot(x=cars.origin, y=cars.horsepower, size=3)  # Reduced size
```

```python
sns.swarmplot(x='origin', y='horsepower',
               data=cars,size=3
              )
```

```python
sns.swarmplot(x='origin', y='horsepower', hue='cylinders',
               data=cars,size=3
              )
```

```python
sns.swarmplot(x='origin', y='horsepower', hue='cylinders',
               data=cars, dodge=True,size=4
              )
```

## Swarmplot Overlay

```python
usa = cars[cars.origin=='usa']
```

### With Boxplot

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower)
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,size=4)
```

```python
sns.boxplot(x=usa.cylinders, y=usa.horsepower,
            whis=np.inf
           )
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='black', alpha=0.5,size=4
             )
```

### With Violinplot

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower)
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,size=4)
```

```python

sns.violinplot(x=usa.cylinders, y=usa.horsepower, density_norm='width')

sns.swarmplot(x=usa.cylinders, y=usa.horsepower, color='white', size=4)
```

```python
sns.violinplot(x=usa.cylinders, y=usa.horsepower,
               density_norm='width', inner=None, color='xkcd:forest green'
              )
sns.swarmplot(x=usa.cylinders, y=usa.horsepower,
              color='white', size=3
             )
```
