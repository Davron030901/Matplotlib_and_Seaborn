# Seaborn: violinplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

## Intro Visuals

```python
import pandas as pd
import numpy as np
```

```python
sns.set_style('white')
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
```

```python
blue, orange, green, red = sns.color_palette()[:4]
```

```python
pts = [1, 1.5, (7/3), 3, 3.75, 4.25, 5, 19/3, 8, 12]
```

```python
pd.Series(pts).describe()
```

### Recipe

```python
plt.figure(figsize=(6,4))
sns.boxplot(pts, color='#44C34A', width=0.5)
sns.despine(left=True)
plt.xlim(-3, 16)
plt.xticks(range(0,13,2))
plt.tight_layout()
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(pts, color='#4BAF50', fill=True, alpha=1, bw_method=1)
sns.kdeplot(pts, color='#7A7A7A', zorder=2, lw=3, bw_method=1)
sns.despine(left=True)
plt.xlim(-3, 16)
plt.ylim(None, 0.3)
plt.xticks(range(0,13,2))
plt.yticks([])
plt.tight_layout()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(pts, color='#44C34A', bw_method=0.3, cut=3)
sns.despine(left=True)
plt.xlim(-3, 16)
plt.xticks(range(0,13,2))
plt.tight_layout()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(pts, color='#44C34A', bw_method=0.3, cut=3, width=0.5)
sns.despine(left=True)
plt.xlim(-3, 16)
plt.xticks(range(0,13,2))
plt.tight_layout()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(pts, color='xkcd:golden yellow', width=0.4)
sns.despine(left=True)
plt.xlim(-3, 16)
plt.xticks(range(0,13,2))
plt.tight_layout()
```

## Basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

Filter down to only cars with even number of cylinders

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(x='cylinders', y='displacement',
               data=cars
              )
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars
              )
```

## Violinplot Options

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])]
              )

plt.legend(loc=2)
```

### split

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True
              )

plt.legend(loc=2)
```

### inner

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               inner='quartiles'
              )

plt.legend(loc=2)
```

### scale

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               density_norm='count'
              )

plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan', 'europe'])],
               split=True,
               density_norm='count',
               common_norm=True
              )

plt.legend(loc=2)
```

```python
(
    cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

## Styling

### Box Options

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               order=['japan', 'europe', 'usa']
              )
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
               linewidth=3
              )
```

### KDE Options

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement,
               bw_method=0.2
              )
```
