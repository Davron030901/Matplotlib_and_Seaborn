```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
colour1=[40,120,60]
colour2=[60,50,90]
ax=plt.figure().add_subplot(projection='3d')
ax.scatter(xs=colour1[0],ys=colour1[1],zs=colour1[2],zdir='z',label='colour1')
ax.scatter(xs=colour2[0],ys=colour2[1],zs=colour2[2],zdir='z',label='colour2')
ax.scatter(xs=0,ys=0,zs=0,zdir='z',label='origin',c='black')
ax.quiver(0,0,0,colour1[0],colour1[1],colour1[2],color='red')
ax.quiver(0,0,0,colour2[0],colour2[1],colour2[2],color='blue')
ax.legend()
ax.set_xlim(0,150)
ax.set_ylim(0,150)
ax.set_zlim(0,150)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(elev=20,azim=-35,roll=0)
plt.show()
```

```python
L1=[colour1[i]-colour2[i] for i in range(len(colour2))]
L1=np.abs(L1).sum()
print(L1)
```

```python
a=np.array(colour1)
b=np.array(colour2)
np.linalg.norm((a-b),ord=1)
```

```python
L2=[(colour1[i]-colour2[i])**2 for i in range(len(colour2))]
L2=np.sqrt(np.array(L2).sum())
print(L2)
```

```python
np.linalg.norm((a-b),ord=2)
```

```python
cosine=np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
print(cosine)
```

```python
dot=np.array([colour1[i]*colour2[i] for i in range(len(colour2))])
dot=dot.sum()
print(dot)
```
