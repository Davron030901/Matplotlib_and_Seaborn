```python
data=sns.load_dataset('diamonds')
data.head()
```

```python
data=data.sample(300)
```

```python
sns.scatterplot(data=data,x='carat',y='price')
```

```python
sns.scatterplot(data=data,x='carat',y='price',markers="+")
```

```python
sns.kdeplot(data=data,x='carat',y='price')
```

```python
sns.kdeplot(data=data,x='carat',y='price',fill=True,cmap='plasma')
```

```python
sns.kdeplot(data=data,x='carat',y='price',fill=True,cmap='mako_r')
```

```python
sns.histplot(data=data,x='carat',y='price')
```

```python
sns.kdeplot(data=data,x='carat',y='price',cmap='coolwarm')
sns.histplot(data=data,x='carat',y='price',color='red')
```

```python
sns.jointplot(data=data,x='carat',y='price')
```

```python
sns.jointplot(data=data,x='carat',y='price',kind='hex')
```

```python
sns.lmplot(data=data,x='carat',y='price',markers=".",line_kws={'color':'red'})
```

```python
sns.relplot(data=data,x='carat',y='price',col='cut')
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
data=sns.load_dataset('diamonds')
data.head()
```

```python
# One-hot encoding
data_encoded = pd.get_dummies(data)
sns.heatmap(data_encoded.corr())
```

```python
sns.heatmap(data_encoded.corr(),annot=True,linewidths=1)
```

```python
sns.heatmap(data_encoded.corr()[['price']],annot=True)
```
