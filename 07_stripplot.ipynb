### Stripplot vs Swarmplot

```python
pts = np.array([1, 1, 1, 1, 1, 2, 2, 3, 4, 4, 4, 4, 5, 5, 7, 7, 7, 8])

plt.figure(figsize=(4, 4))
sns.stripplot(x=pts, color=green, size=15, linewidth=1, edgecolor='black', alpha=0.7)
sns.despine(left=True)
plt.xlim(0, 9)
plt.tight_layout()
```

```python
pts = np.array([1, 1, 1, 1, 1, 2, 2, 3, 4, 4, 4, 4, 5, 5, 7, 7, 7, 8])

plt.figure(figsize=(4, 4))
sns.swarmplot(x=pts, color=green, size=15, linewidth=1, edgecolor='black')
sns.despine(left=True)
plt.xlim(0, 9)
plt.tight_layout()
```

## Example Gallery

```python
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('axes', labelsize=14)
```

#### Cylinders by hue

```python
sns.set_style('white')
sns.stripplot(x=cars.weight, y=['']*len(cars), hue=cars.cylinders,
              palette=sns.cubehelix_palette(3))
sns.despine(left=True)
plt.xlabel('\nWeight (lbs)')
plt.legend(title='Cylinders', frameon=False)
```

#### Newer models, vertical

```python
sns.set_style('darkgrid')
cars['newer_model'] = [True if mod_year > 76 else False for mod_year in cars.model_year]
```

```python
sns.stripplot(x=cars.newer_model, y=cars.weight,
              hue=cars.newer_model, jitter=0.2,
              palette=['gray', 'red'], legend=False)
plt.xlabel('\nNewer Model (post-1976)')
plt.ylabel('Weight (lbs)', rotation=0, labelpad=50)
plt.tight_layout()
```

#### Low, medium, high MPG

```python
cars['mpg_range'] = ['Low' if m < 19 else ('Medium' if m < 27 else 'High') for m in cars.mpg]
```

```python
sns.set_style('whitegrid')
sns.stripplot(x=cars.weight, y=cars.mpg_range,
              hue=cars.mpg_range,  # Add hue parameter
              order=['High', 'Medium', 'Low'],
              jitter=0.2,
              palette=["#58b5e1", "#0db293", "#19282d"],
              alpha=0.5,
              size=8,
              legend=False  # Add legend=False
             )
plt.xlabel('\nWeight (lbs)', color='gray')
plt.ylabel('Miles per\ngallon (mpg)',
           rotation=0,
           labelpad=80,
           ha='left',
           position=(-.1, .8),
           color='gray'
          )
plt.tight_layout()
```

#### Company smileys

```python
import random
random.seed(11)

points = 50
data = [random.gauss(28, 5) for _ in range(points)]
data.extend([random.gauss(20, 8) for _ in range(points)])
data.extend([random.gauss(15, 5) for _ in range(points)])
company = ['Company A']*points + ['Company B']*points + ['Company C']*points

plt.figure(figsize=(8,5))
sns.stripplot(x=data, y=company,
              hue=company,  # Add hue parameter
              jitter=0.2,
              alpha=0.6,
              size=20,
              marker=r'$$',
              palette=["#750e3d", "#d34069", "#190e11"],
              legend=False  # Add legend=False
             )
for pos in ['left', 'right', 'top', 'bottom']:
    plt.gca().spines[pos].set_edgecolor('lightgray')
plt.xticks([0, 10, 20, 30, 40, 50])
plt.tight_layout()
```

#### Boxplot overlay

```python
sns.boxplot(
    x=cars.weight,
    y=cars.origin,
    whis=0,
    color='white',
    fliersize=0,
    order=['usa', 'europe', 'japan']
)

# Stripplot for individual data points
sns.stripplot(
    x=cars.weight,
    y=cars.origin,
    hue=cars.origin,  # Assign the categorical variable to hue
    palette=["#0f1745", "#2580fe", "#3123a1"],
    order=['usa', 'europe', 'japan'],
    legend=False  # Disable legend since it's not needed
)

# Customize the plot
plt.xlabel('\nWeight (lbs)')
plt.ylabel('')  # Remove y-axis label
plt.yticks([0, 1, 2], ['USA', 'Europe', 'Japan'])  # Customize y-ticks
plt.tight_layout()

# Show the plot
plt.show()
```
