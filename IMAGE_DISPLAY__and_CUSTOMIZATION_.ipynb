#Random Image - RGB! üåà

````python
np.random.seed(7)
image = np.random.rand(200, 200, 3)

print(f"Image shape: {image.shape}")
print(f"Data type: {image.dtype}")
print(f"Value range: [{image.min():.2f}, {image.max():.2f}]")
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.title("Random RGB Image", fontsize=16)
plt.axis("off")
plt.show()
````

#Grayscale + Heatmap! üî•

````python
np.random.seed(100)
data = np.random.randn(50, 50)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

im1 = axes[0].imshow(data, cmap="gray")
axes[0].set_title("Grayscale", fontsize=14)
axes[0].axis('off')
plt.colorbar(im1, ax=axes[0], label="Intensity")

im2 = axes[1].imshow(data, cmap='hot')
axes[1].set_title("Heatmap", fontsize=14)
axes[1].axis('off')
plt.colorbar(im2, ax=axes[1], label='Temperature')

plt.tight_layout()
plt.show()
````

#Correlation Matrix - Real Data! üìä

````python
np.random.seed(9)
math = np.random.normal(70, 15, 100)
physics = math * 0.8 + np.random.randn(100) * 10
english = np.random.normal(75, 10, 100)
chemistry = physics * 0.7 +np.random.randn(100) * 12
biology = english * 0.6 + np.random.randn(100) * 15

data = np.column_stack([math, physics, english, chemistry, biology])

correlation = np.corrcoef(data.T)

print("Subjects: `Math`, `Physics`, `English`, `Chemistry`, `Biology`")
print("Correlation Matrix shape:", correlation.shape)
print("Correlation Matrix:")
print(correlation)
````

````python
plt.figure(figsize=(10, 8))
im = plt.imshow(correlation, cmap="coolwarm", vmin=-1, vmax=1)

subjects = ['Math', 'Physics', 'English', 'Chemistry', 'Biology']
plt.xticks(range(5), subjects, rotation=45)
plt.yticks(range(5), subjects)

plt.colorbar(im, label="Correlation")

for i in range(5):
  for j in range(5):
    text = plt.text(j, i, f"{correlation[i, j]:.2f}",
                    ha='center', va='center',
                    color='white' if abs(correlation[i, j]) > 0.5 else 'black')

plt.title("Subject Correlation Matrix", fontsize=16)
plt.tight_layout()
plt.show()
````

#Different Colormaps Comparison! üé®

````python
np.random.seed(2)
data = np.random.randn(20, 20)

colormaps = ['viridis', 'hot', 'cool', 'rainbow', 'gray', 'seismic']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()
for idx, cmap_name in enumerate(colormaps):
  im = axes[idx].imshow(data, cmap=cmap_name)
  axes[idx].set_title(f'{cmap_name}', fontsize=14)
  axes[idx].axis('off')
  plt.colorbar(im, ax=axes[idx])

plt.suptitle("Colormap Comparison", fontsize=18)
plt.tight_layout()
plt.show()
````

#Gradient Image

````python
gradient_image = np.linspace(0, 1, 300).reshape(-1, 1) * np.ones((1, 300))
print(f"Value range: [{gradient_image.min():.2f}, {gradient_image.max():.2f}]")
````

````python
plt.figure(figsize=(8, 8))
plt.imshow(gradient_image, cmap="viridis")
plt.colorbar(label='Intensity')
plt.title("Vertical Gradient")
````

#Temperature Heatmap

````python
temperature = np.random.uniform(15, 35, (7, 24))

plt.imshow(temperature, cmap='hot')
plt.xticks(np.arange(24), rotation=270)
plt.yticks(np.arange(7))
plt.colorbar(label='Temperature')
plt.title("Temperature Heatmap")
plt.xlabel("Hour")
plt.ylabel("Day")
plt.show()
````

#Multi-Colormap Display

````python
data = np.random.randn(50, 50)
colormaps = ['gray', 'hot', 'cool', 'rainbow']
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
for idx, cmap_name in enumerate(colormaps):
  im = axes[idx%2, idx//2].imshow(data, cmap=cmap_name)
  axes[idx%2, idx//2].set_title(f'{cmap_name}', fontsize=14)
  axes[idx%2, idx//2].axis('off')
  plt.colorbar(im, ax=axes[idx%2, idx//2])

plt.suptitle("Colormap Comparison", fontsize=18)
plt.tight_layout()
plt.show()
````
