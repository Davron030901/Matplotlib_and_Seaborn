# Seaborn: catplot

```python
import seaborn as sns
```

```python
#load car data and drop rows with missings
cars = sns.load_dataset('mpg').dropna()

#filter to 4-, 6-, 8-cylinder cars
cars = cars[cars.cylinders.isin([4, 6, 8])]

#add old/new column
cars['type'] = ['old' if x <=76 else 'new' for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

## Intro Visuals

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
sns.set(style='darkgrid', font_scale=1.5)

g = sns.catplot(x='cylinders', y='horsepower',
                data=cars,
                row='origin',
                kind='box',
                aspect=1.5,
                height=4
               )
g.set_titles(row_template="{row_name} cars")
```

### Thumbnail

```python
sns.set(style='white', font_scale=1.5)

g = sns.catplot(x='cylinders', y='mpg',
                data=cars,
                col='type',
                kind='violin',
                hue='cylinders',
                aspect=1.7,
                height=4,
                facet_kws={'ylim':(0, 50)},
                palette=['lightgray', '#B58389', '#7C2021']
               )
g.set_titles(col_template="{col_name} cars")
plt.tight_layout()
```

```python

sns.set(style='white', font_scale=1.5)

g = sns.catplot(x='horsepower', y=['']*len(cars),
                hue='cylinders',
                data=cars,
                kind='strip',
                aspect=4.8,
                height=3,
                s=10,
                legend=False,
                palette=['lightgray', '#B58389', '#7C2021'],
               )
g.set_titles(col_template="{col_name} cars")
plt.tight_layout()
```

## Small Multiples (FacetGrid)

```python
sns.set_style('darkgrid')
```

#### `row`, `col`

```python
sns.catplot(x='cylinders', y='horsepower',
            data=cars,
            kind='box',
            col='origin'
           )
```

```python
sns.catplot(x='cylinders', y='horsepower',
            data=cars,
            kind='strip',
            col='origin'
           )
```

```python
sns.catplot(x='cylinders', y='horsepower',
            data=cars,
            kind='strip',
            col='origin',
            row='type'
           )
```

#### `row_order` (`order`, `hue_order`, `col_order`)

```python
sns.catplot(x='cylinders', y='horsepower',
            data=cars,
            kind='strip',
            col='origin',
            row='type',
            row_order=['new', 'old']
           )
```

## Styling

In general, additional keyword arguments will be passed to the kind of plot selected.  

```python
sns.catplot(x='cylinders', y='horsepower',
            data=cars,
            kind='box',
            col='origin',
            hue='cylinders',
            palette='winter',
            whis=1
           )
```

```python
sns.catplot(x='cylinders', y='horsepower',
            data=cars,
            kind='strip',
            col='origin',
            s=12,
            alpha=0.3,
            linewidth=2
           )
```
